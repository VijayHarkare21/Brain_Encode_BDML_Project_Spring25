```python
import pickle as pkl
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
results_dir = Path("/scratch/vjh9526/bdml_2025/project/code/classifiers/results/unorganized")
output_file = Path("/scratch/vjh9526/bdml_2025/project/code/classifiers/eeg_cls_all_results.csv")
```

```python
rows = []

for csv_path in sorted(results_dir.glob("*_summary.csv")):
    df = pd.read_csv(csv_path)

    # Each summary file has a single row
    cls      = df.loc[0, "classifier"]
    accuracy = df.loc[0, "accuracy"]

    # Drop the trailing "_summary.csv" to get the embed-model identifier
    complete_model = csv_path.stem.rsplit("_summary", 1)[0]
    model_desc = complete_model.split("_")
    embed_model = model_desc[0]
    modality = model_desc[1]
    mode = model_desc[2]

    rows.append({"embed_model": embed_model,
                 "modality": modality,
                 "mode": mode,
                 "cls":         cls,
                 "accuracy":    accuracy})

# Combine and save
pd.DataFrame(rows).to_csv(output_file, index=False)
print(f"Saved aggregated results to {output_file.resolve()}")
```
