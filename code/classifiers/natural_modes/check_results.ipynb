```python
import pickle as pkl
import os
import numpy as np
import pandas as pd
```

```python
results_dir = "/scratch/vjh9526/bdml_2025/project/code/classifiers/natural_modes"
```

```python
with open(os.path.join(results_dir, "results_new.pkl"), mode="rb") as f:
    results_imgs = pkl.load(f)
```

```python
results_imgs

results_dict = {
    "embed_model": [],
    "cls": [],
    "accuracy": [],
    "precision": [],
    "recall": [],
    "f1": [],
}

for embed_model in results_imgs:
    for cls in results_imgs[embed_model]:
        results_dict["embed_model"].append(embed_model)
        results_dict["cls"].append(cls)
        results_dict["accuracy"].append(results_imgs[embed_model][cls]["accuracy"])
        results_dict["precision"].append(results_imgs[embed_model][cls]["precision"])
        results_dict["recall"].append(results_imgs[embed_model][cls]["recall"])
        results_dict["f1"].append(results_imgs[embed_model][cls]["f1"])
```

```python
results_dict
```

```python
df = pd.DataFrame(results_dict)
df
```

```python
df.to_csv(os.path.join(results_dir, "results_image_cls.csv"))
```
